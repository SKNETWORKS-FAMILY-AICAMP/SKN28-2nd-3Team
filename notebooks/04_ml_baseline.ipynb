# 04. 머신러닝 베이스라인 모델 학습

이 단계에서는 고객 이탈 예측 문제에 대해 기본적인 머신러닝 모델을 적용하여  
초기 성능 기준선(baseline)을 설정한다.

베이스라인 모델은 이후 고도화된 모델과 비교하기 위한 기준점 역할을 하며,  
데이터만으로 어느 정도 수준까지 예측이 가능한지를 빠르게 확인하는 데 목적이 있다.

일반적으로 이 단계에서는 다음과 같은 작업을 수행한다.

- 학습용 데이터(train)와 평가용 데이터(test) 분리
- 여러 머신러닝 알고리즘 적용 (예: Logistic Regression, Random Forest, Gradient Boosting 등)
- 기본 하이퍼파라미터로 모델 학습
- 예측 성능 평가 (Accuracy, Precision, Recall, F1-score 등)

특히 고객 이탈 예측에서는 단순 정확도(Accuracy)보다  
이탈 고객을 얼마나 잘 잡아내는지(Recall),  
불필요한 오탐을 얼마나 줄이는지(Precision)가 중요하게 고려된다.

이 단계는 복잡한 모델을 적용하기 전에  
문제의 난이도와 데이터의 예측 가능성을 파악하는 출발점이라고 볼 수 있다.
