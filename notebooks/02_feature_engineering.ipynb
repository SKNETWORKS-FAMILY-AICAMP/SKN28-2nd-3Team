# 02. 피처 엔지니어링 및 학습용 테이블 생성

이 단계에서는 원천 데이터를 바로 모델에 넣는 대신,  
예측에 활용할 수 있는 형태로 가공하여 학습용 테이블을 구성한다.

실제 비즈니스 데이터는 하나의 깔끔한 테이블로 바로 주어지지 않는 경우가 많다.  
이 프로젝트 역시 accounts, subscriptions, feature_usage, support_tickets 등  
서로 다른 테이블이 존재하며, 각 테이블은 행의 기준 단위가 서로 다르다.

따라서 이 단계의 핵심은  
여러 테이블의 정보를 실제 예측 단위인 account 수준으로 통합하고,  
모델이 학습할 수 있는 입력 변수로 변환하는 것이다.

이 과정에서 수행하는 작업은 보통 다음과 같다.

- 테이블 간 조인 및 기준 단위 통일
- 집계 변수 생성
- 기간 기반 파생 변수 생성
- 결측치 처리 및 기본 정제
- 학습에 사용할 최종 입력 테이블 구성

즉, 피처 엔지니어링은 단순히 변수를 추가하는 작업이 아니라,  
원천 데이터를 예측 가능한 구조로 재해석하는 과정이라고 볼 수 있다.
