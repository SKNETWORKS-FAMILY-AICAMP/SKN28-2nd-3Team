# 03. 탐색적 데이터 분석

이 단계에서는 피처 엔지니어링을 거쳐 생성된 학습용 테이블을 바탕으로  
데이터의 전반적인 분포와 특성을 탐색한다.

모델 성능을 높이기 위해서는 단순히 알고리즘을 적용하는 것보다,  
먼저 데이터가 어떤 구조를 가지고 있는지 이해하는 과정이 중요하다.  
특히 고객 이탈 예측 문제에서는 이탈 여부에 따라 어떤 변수 차이가 나타나는지,  
어떤 변수에서 불균형이나 왜곡이 존재하는지를 미리 확인할 필요가 있다.

이 단계에서 중점적으로 살펴보는 내용은 다음과 같다.

- 타깃 변수의 분포와 클래스 불균형 여부
- 수치형 변수의 분포와 이상치 존재 여부
- 범주형 변수의 구성과 주요 비율
- 이탈 고객과 비이탈 고객 간의 특성 차이
- 변수별 전반적 패턴과 해석 가능성

즉, 탐색적 데이터 분석은 단순한 시각화 작업이 아니라,  
데이터를 해석하고 이후 모델링 방향을 결정하기 위한 이해의 단계라고 볼 수 있다.
